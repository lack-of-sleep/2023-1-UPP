```python
#02
from __future__ import division
def lexical_diversity(text):
  return len(text) / len(set(text))
```

```python
#03
def lexical_diversity(my_text_data):
  word_count = len(my_text_data) # word tokens
  vocab_size = len(set(my_text_data)) # word types
  diversity_score = vocab_size / word_count # Type/Token Ratio (TTR)
  return diversity_score
```

```python
#04
from nltk.corpus import genesis
kjv = genesis.words('english-kjv.txt')
lexical_diversity(kjv)
```

```python
#05
def plural(word):
  if word.endswith('y'): # if
    return word[:-1] + 'ies'
  elif word[-1] in 'sx' or word[-2:] in ['sh', 'ch']: # else if
    return word + 'es'
  elif word.endswith('an'):
    return word[:-2] + 'en'
  else: # else
    return word + 's'
```

```python
#06
plural('fairy')
```

```python
plural('woman')
```

```python
#08
def unusual_words(text):
  text_vocab = set(w.lower() for w in text if w.isalpha())
  english_vocab = set(w.lower() for w in nltk.corpus.words.words())
  unusual = text_vocab - english_vocab
  return sorted(unusual)
```

```python
#09
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

```python
#11
def content_fraction(text):
  stopwords = nltk.corpus.stopwords.words('english') # get the lsit of stopwords
  content = [w for w in text if w.lower() not in stopwords] # eliminate the stopwords
  return len(content) / len(text)
content_fraction(nltk.corpus.reuters.words())
```

```python
#20
p3 = [(pron[0]+'-'+pron[2], word)
  for (word, pron) in entries
  if pron[0] == 'P' and len(pron) == 3]

cfd = nltk.ConditionalFreqDist(p3)

for template in sorted(cfd.conditions()):
  if len(cfd[template]) > 10:
    words = sorted(cfd[template])
    wordstring = ' '.join(words)
    print(template, wordstring[:70] + "...")
```
